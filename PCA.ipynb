```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df_iris = pd.DataFrame(data=iris.data,columns=iris.feature_names)

df_iris['label'] = iris.target

df_iris.head(3)
```

```python
import matplotlib.pyplot as plt
markers = ['^', 's', 'o']

for i, marker in enumerate(markers):
    x_axis_data = df_iris[df_iris['label'] == i]['petal length (cm)']
    y_axis_data = df_iris[df_iris['label'] == i]['petal width (cm)']
    plt.scatter(x_axis_data, y_axis_data, marker = marker, label = iris.target_names[i])

plt.legend()
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(iris.data)
```

```python
df_scaled_iris = pd.DataFrame(data=X_scaled,columns=iris.feature_names)
df_scaled_iris['label'] = iris.target
```

```python
df_scaled_iris.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X_scaled)

iris_features_pca = pca.transform(X_scaled)
```

```python
pca.components_
```

```python
pca.singular_values_
```

```python
pca.explained_variance_ratio_
```

```python
0.72962445+0.22850762
```

```python
pca_columns = ['pca_component_1', 'pca_component_2']

pca_df = pd.DataFrame(iris_features_pca, columns=pca_columns)
pca_df['target'] = iris.target
pca_df.head(3)
```

```python
import matplotlib.pyplot as plt
markers = ['^', 's', 'o']

for i, marker in enumerate(markers):
    x_axis_data = pca_df[pca_df['target'] == i]['pca_component_1']
    y_axis_data = pca_df[pca_df['target'] == i]['pca_component_2']
    plt.scatter(x_axis_data, y_axis_data, marker = marker, label = iris.target_names[i])

plt.legend()
plt.xlabel('pca_component_1')
plt.ylabel('pca_component_2')
plt.show()
```
