```python
#추천시스템 class

from sklearn.metrics.pairwise import cosine_similarity
import random
average_series = ratings_df.groupby('movieId')['rating'].mean()
count_series = ratings_df.groupby('movieId')['rating'].count()


df_1 = average_series.reset_index()
df_2 = count_series.reset_index()
df_3 = df_1.rename(columns={'rating':'average_rating'})
df_4 = df_2.rename(columns={'rating':'rating_count'})
df = pd.merge(df_3, df_4 , on='movieId')
merge_df = pd.merge(movies_df, df , on='movieId')

# 가중평점 계산해서 weighted_rating 컬럼 추가
C = merge_df['average_rating'].mean()
# 투표횟수 중 60%이상의 횟수에 달하는 숫자
# 예를들어 총 투표횟수가 100과 1일때 m값은 매우 달라진다.
m = merge_df['rating_count'].quantile(0.6)

def weighted_vote_average(record):
    v = record['rating_count']
    R = record['average_rating']
    
    return ( (v/(v+m)) * R) + ( (m/(m+v)) * C)

merge_df['weighted_rating'] = merge_df.apply(weighted_vote_average, axis=1)


class GET_REMMAND():
    def __init__(self):
        # timestamp 제거
        self.movie_ratings_df = pd.merge(movies_df, ratings_df, on="movieId")
        self.user_movie_rating_df = pd.pivot_table(movie_ratings_df, values='rating', index=['userId'],
                       columns=['title'])
        # null을 0을 처리
        self.user_movie_rating_df = user_movie_rating_df.fillna(0)

    def cosine_similarity(self , based="user_based"):
        if based == "user_based":
            #user 코사인 유사도 
            cosine_sim = cosine_similarity(self.user_movie_rating_df , self.user_movie_rating_df)
            return cosine_sim
        elif based == "content_based":
            pass
            
    # 비슷한 유저에서 타겟 유저간 안본 영화
    def get_user_based_movies(self ,target_user_id, cosine_sim ,  similar_user=5 ):
        # 타겟유저와 자기 자신으로 제외한 비슷한 유저 5명
        user_sim_sorted_ind = cosine_sim.argsort()[: , ::-1]
        user_sim_sorted_ind[target_user_id][1:similar_user+1]
        # 타겟유저가 본영화
        target_user_rating_list = list(user_movie_rating_df.iloc[target_user_id , : ])
        target_user_movie = [x[0] for x in enumerate(target_user_rating_list) if x[1] !=0]

        # 유사 유저가 본영화
        other_user_movie = []
        for i in user_sim_sorted_ind[target_user_id][1:6]:
            result = [x[0] for x in enumerate(list(user_movie_rating_df.iloc[i , :])) if x[1] !=0 ]
            other_user_movie += result
        
        result_movie_index_list = list(set(other_user_movie) - set(target_user_movie))

        return result_movie_index_list

    # 가중 평점이 높은 영화를 n개 추천
    def weighted_ratings_movie_remmand(self ,n=10):
        movies_series = pd.Series(movies_df['title'] , index=movies_df.index)

        topn_movies = merge_df.iloc[result_movie_index_list , :].sort_values(by=['weighted_rating'] ,ascending=False)[:n]['title']
        return topn_movies

    # 비슷한 유저가 본영화중에서 타겟유저가 본 영화의 콘텐츠가 비슷한 영화 n개 추천
    def target_user_movie_recommand(self , target_user , target_movie , result_movie_index_list, n=10):
        return result_movie_index_list
        
        
        
        
        
```
