```python
import pandas as pd
def get_recommander(title , cosine_sim = soup_sim_sorted_ind , limit_idx=20):
    indices = pd.Series(movies_5000.index , index= movies_5000['original_title'])
    idx = indices[title]
    movie_indices = cosine_sim[idx][1:limit_idx+1]
    return movies_5000['original_title'].iloc[movie_indices]
```

```python
# 가중평점이 좋은영화 유사한 영화 추천
C = movies_5000['vote_average'].mean()
# 투표횟수 중 60%이상의 횟수에 달하는 숫자
# 예를들어 총 투표횟수가 100과 1일때 m값은 매우 달라진다.
m = movies_5000['vote_count'].quantile(0.6)
def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']
    
    return ( (v/(v+m)) * R) + ( (m/(m+v)) * C)
    
movies_5000['weighted_vote'] = movies_5000.apply(weighted_vote_average, axis=1)
```

```python
movies_5000.head(3)
```

```python
import pandas as pd
def get_weighted_vote_recommander(title , cosine_sim = soup_sim_sorted_ind , limit_idx=20):
    indices = pd.Series(movies_5000.index , index= movies_5000['original_title'])
    idx = indices[title]
    movie_indices = cosine_sim[idx][1:limit_idx+1]
    return movies_5000.iloc[movie_indices].sort_values(by=['weighted_vote'] , ascending=False)['title']
```

```python
get_weighted_vote_recommander(title='Avatar', limit_idx=10)
```

```python
title = input("당신이 좋아하는 영화를 입력하시오!!")
result = get_weighted_vote_recommander(title=title , limit_idx=10)
print(result)
```

```python
import pandas as pd
import numpy as np

movies_df = pd.read_csv('./datasets/movies_small.csv')
ratings_df = pd.read_csv('./datasets/ratings_small.csv')
```

```python
# timestamp 제거
ratings_df.drop('timestamp' ,axis=1, inplace=True)
```

```python
movie_ratings_df = pd.merge(movies_df, ratings_df, on="movieId")
```

```python
movie_ratings_df.info()
```

```python
movie_ratings_df.head()
```

```python
user_movie_rating_df = pd.pivot_table(movie_ratings_df, values='rating', index=['userId'],
                       columns=['title'])
```

```python
user_movie_rating_df.iloc[0, :]
```

```python
user_movie_rating_df.head()
```

```python
# null을 0을 처리
user_movie_rating_df = user_movie_rating_df.fillna(0)
```

```python
user_movie_rating_df.head()
```

```python
#user 코사인 유사도 
user_cosine_sim = cosine_similarity(user_movie_rating_df , user_movie_rating_df)
```

```python
list(enumerate(user_cosine_sim[0]))
```

```python
user_sim_sorted_ind = user_cosine_sim.argsort()[: , ::-1]
user_sim_sorted_ind[0][1:6]
```

```python
# 타겟유저가 보지않은 영화를 비슷한 유저가 본영화중에 추천
# 타겟유저가 본영화 
taget_user = 0
target_user_rating_list = list(user_movie_rating_df.iloc[taget_user , : ])
```

```python
target_user_movie = [x[0] for x in enumerate(target_user_rating_list) if x[1] !=0]
```

```python
# 유사 유저가 본영화
other_user_movie = []
taget_user = 0
for i in user_sim_sorted_ind[taget_user][1:6]:
    result = [x[0] for x in enumerate(list(user_movie_rating_df.iloc[i , :])) if x[1] !=0 ]
    other_user_movie += result
list(set(other_user_movie))
```

```python
import random
result_movie_index_list = list(set(list(set(other_user_movie))) - set(target_user_movie))

# 리스트에서 2개 랜덤 추출

random_movie_list = random.sample(result_movie_index_list, 10)
```

```python
movies_series = pd.Series(movies_df['title'] , index=movies_df.index)
movies_series
```
